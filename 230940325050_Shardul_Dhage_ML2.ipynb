```python
X=df[['V1', 'V3', 'V4', 'V9', 'V10', 'V11', 'V12', 'V14', 'V16']]
Y=df.iloc[:,-1]
```

```python
# Problem St.:
#  We have to Develop a machine learning model to accurately identify fraudulent credit card transactions among
# a dataset of European cardholder transactions from the year 2023.
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
classifier=lr.fit(X_train,Y_train)
```

```python
Y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
accuracy_score(Y_pred,Y_test)
```

```python
confusion_matrix(Y_pred,Y_test)
```

```python
print(classification_report(Y_pred,Y_test))
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
classifier1=dt.fit(X_train,Y_train)
```

```python
Y_pred1=dt.predict(X_test)
```

```python
from sklearn import tree
text=tree.export_text(classifier1)
print(text)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
accuracy_score(Y_pred1,Y_test)
```

```python
confusion_matrix(Y_pred1,Y_test)
```

```python
print(classification_report(Y_pred1,Y_test))
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100,criterion='entropy',random_state=0)
classifier2=rf.fit(X_train,Y_train)
```

```python
Y_pred2=rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
accuracy_score(Y_pred2,Y_test)
```

```python
confusion_matrix(Y_pred2,Y_test)
```

```python
print(classification_report(Y_pred2,Y_test))
```

```python
# Gradient Boosting
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
# Create a Gradient Boosting Classifier
gb_classifier = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=42)

# Train the model
gb_classifier.fit(X_train, Y_train)

# Make predictions
Y_pred3 = gb_classifier.predict(X_test)
```

```python
confusion_matrix(Y_pred3, Y_test)
```

```python
# Evaluate accuracy
accuracy = accuracy_score(Y_pred3,Y_test)
accuracy
```

```python
# Display additional metrics
print(classification_report(Y_pred3, Y_test))
```

```python
#SVM(Non Linear)
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix

# Create an SVM model with RBF kernel
svm_model = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)
# Other kernel options: 'poly', 'sigmoid', etc.

# Train the model
svm_model.fit(X_train, Y_train)

# Make predictions
Y_pred4 = svm_model.predict(X_test)

# Display additional metrics
print(classification_report(Y_test, Y_pred4))
```

```python
# Evaluate accuracy
accuracy = accuracy_score(Y_test,Y_pred4 )
print(f"Accuracy: {accuracy}")
```

```python
#Initializing the ANN
import tensorflow as tf
ann=tf.keras.models.Sequential()
```

```python
from keras import layers
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
#adding input layer
ann.add(Dense(units=164, activation='relu'))
```

```python
#adding second hidden layer
ann.add(Dense(units=64, activation='relu'))
```

```python
ann.add(Dense(units=28, activation='relu'))
```

```python
#adding an output layer
ann.add(Dense(units=1, activation='sigmoid'))
```

```python
ann.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#fitting
ann.fit(X_train,Y_train,batch_size=1000,epochs=3)
```

```python
#prediction
Y_pred5=ann.predict(X_test)
```

```python
Y_pred5
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# Assuming y_pred contains continuous values
threshold = 0.5
y_pred_binary = (Y_pred5 > threshold).astype(int)

# Now you can use the classification metrics with y_test and y_pred_binary
print("Confusion Matrix:\n", confusion_matrix(Y_test, y_pred_binary))
print("\nClassification Report:\n", classification_report(Y_test, y_pred_binary))
print("\nAccuracy Score:", accuracy_score(Y_test, y_pred_binary) * 100, "%")
```
